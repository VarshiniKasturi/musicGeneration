```python
import os
import pretty_midi

# Function to extract musical information from MIDI files
def extract_midi_info(midi_file_path):
    try:
        # Load MIDI file using pretty_midi
        midi_data = pretty_midi.PrettyMIDI(midi_file_path)

        # Extract notes, timing, and velocities
        notes = []
        for instrument in midi_data.instruments:
            for note in instrument.notes:
                notes.append({
                    'pitch': note.pitch,
                    'start_time': note.start,
                    'end_time': note.end,
                    'velocity': note.velocity
                })

        return notes
    except Exception as e:
        print(f"Error processing MIDI file '{midi_file_path}': {e}")
        return []

# Set the path to the dataset folder in Google Drive
dataset_folder_path = '/content/drive/MyDrive/Colab Notebooks/dataset'

# Iterate through each folder in the dataset
all_notes = []
for folder_name in os.listdir(dataset_folder_path):
    folder_path = os.path.join(dataset_folder_path, folder_name)
    if os.path.isdir(folder_path):
        # List MIDI files in the folder
        midi_files = [file for file in os.listdir(folder_path) if file.endswith('.midi') or file.endswith('.mid')]
        # Process each MIDI file
        for midi_file in midi_files:
            midi_file_path = os.path.join(folder_path, midi_file)
            notes = extract_midi_info(midi_file_path)
            all_notes.extend(notes)

# Print the total number of notes extracted
print(f"Total number of notes extracted: {len(all_notes)}")
import numpy as np

# Function to preprocess the extracted musical information
def preprocess_data(notes, sequence_length=100):
    # Create a list to store input sequences and output targets
    input_sequences = []
    output_targets = []

    # Iterate through the notes and create sequences
    for i in range(0, len(notes) - sequence_length, 1):
        # Extract input sequence and output target
        input_seq = notes[i:i + sequence_length]
        output_target = notes[i + sequence_length]

        # Append the sequences to the lists
        input_sequences.append(input_seq)
        output_targets.append(output_target)

    # Convert the sequences to numpy arrays
    X = np.array(input_sequences)
    y = np.array(output_targets)

    return X, y

# Preprocess the extracted musical information
sequence_length = 100  # Define the length of input sequences
X, y = preprocess_data(all_notes, sequence_length=sequence_length)

# Print the shape of the preprocessed data
print("Shape of input sequences (X):", X.shape)
print("Shape of output targets (y):", y.shape)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding

# Define the model architecture
def build_model(input_shape, vocab_size):
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=256, input_shape=input_shape),
        LSTM(units=1024, return_sequences=True),
        Dense(units=vocab_size)
    ])
    return model

# Define the vocabulary size
vocab_size = 128  # Example vocabulary size, adjust as needed
sequence_length = 100
# Build the model
model = build_model(input_shape=(sequence_length,), vocab_size=vocab_size)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

# Print the model summary
model.summary()

# Define training parameters
batch_size = 64
epochs = 10  # Adjust as needed

# Train the model
# history = model.fit(X, y, batch_size=batch_size, epochs=epochs, validation_split=0.2)

# Evaluate the trained model
# (Optional: You can evaluate the model's performance on a separate validation set if available)
# loss = model.evaluate(X_val, y_val)
# Train the model
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_val, y_val))

# Evaluate the trained model
loss = model.evaluate(X_test, y_test)
# Save the trained model
model.save("music_generation_model.h5")
```
